```python
# =============================================================================
# ПОЛНЫЙ КОД ДЛЯ АНАЛИЗА ДАННЫХ В GOOGLE COLAB
# =============================================================================

# Импорт необходимых библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Настройка отображения графиков
%matplotlib inline
plt.style.use('seaborn-v0_8')

print("=== АНАЛИЗ ДАННЫХ В GOOGLE COLAB ===\n")

# =============================================================================
# 1. ЗАГРУЗКА ДАННЫХ
# =============================================================================

print("1. ЗАГРУЗКА ДАННЫХ...")

# Способ 1: Загрузка с локального компьютера
print("Загрузите файл данных...")
uploaded = files.upload()

# Получаем имя загруженного файла
file_name = list(uploaded.keys())[0]
print(f"Загружен файл: {file_name}")

# Чтение данных
df = pd.read_csv(file_name)

print("Данные успешно загружены!\n")

# =============================================================================
# 2. ПЕРВИЧНЫЙ ОСМОТР ДАННЫХ
# =============================================================================

print("2. ПЕРВИЧНЫЙ ОСМОТР ДАННЫХ:")

# Просмотр первых 5 строк
print("\nПервые 5 строк данных:")
display(df.head())

# Основная информация о данных
print("\nИнформация о данных:")
df.info()

# Основные статистические характеристики
print("\nСтатистические характеристики числовых данных:")
display(df.describe())

# Размерность данных
print(f"\nРазмерность данных: {df.shape}")
print(f"Количество строк: {df.shape[0]}")
print(f"Количество столбцов: {df.shape[1]}")

# =============================================================================
# 3. ПРОВЕРКА ТИПОВ ДАННЫХ И ПРОПУСКОВ
# =============================================================================

print("\n3. ПРОВЕРКА ТИПОВ ДАННЫХ И ПРОПУСКОВ:")

# Проверка типов данных
print("\nТипы данных каждого столбца:")
print(df.dtypes)

# Проверка пропущенных значений
print("\nПропущенные значения:")
missing_data = df.isnull().sum()
print(missing_data)

# Визуализация пропущенных значений
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), yticklabels=False, cbar=True, cmap='viridis')
plt.title('Визуализация пропущенных значений')
plt.show()

# Процент пропущенных значений
print("\nПроцент пропущенных значений:")
missing_percent = (df.isnull().sum() / len(df)) * 100
print(missing_percent)

# =============================================================================
# 4. ПРЕОБРАЗОВАНИЕ ТИПОВ ДАННЫХ
# =============================================================================

print("\n4. АНАЛИЗ КАТЕГОРИАЛЬНЫХ ПЕРЕМЕННЫХ:")

# Просмотр уникальных значений в категориальных столбцах
categorical_columns = df.select_dtypes(include=['object']).columns

if len(categorical_columns) > 0:
    print("Категориальные столбцы и их уникальные значения:")
    for col in categorical_columns:
        print(f"{col}: {df[col].unique()} (всего: {len(df[col].unique())} значений)")
else:
    print("Категориальные столбцы не найдены")

# =============================================================================
# 5. ВИЗУАЛИЗАЦИИ
# =============================================================================

print("\n5. ПОСТРОЕНИЕ ВИЗУАЛИЗАЦИЙ...")

# Визуализация 1: Распределение числовых переменных
numerical_columns = df.select_dtypes(include=[np.number]).columns

if len(numerical_columns) > 0:
    print("\nВизуализация 1: Распределение числовых переменных")
    plt.figure(figsize=(15, 10))
    for i, col in enumerate(numerical_columns, 1):
        plt.subplot(2, 3, i) if len(numerical_columns) > 3 else plt.subplot(1, len(numerical_columns), i)
        df[col].hist(bins=30, alpha=0.7, color='skyblue')
        plt.title(f'Распределение {col}')
        plt.xlabel(col)
        plt.ylabel('Частота')
    plt.tight_layout()
    plt.show()
else:
    print("Числовые столбцы не найдены")

# Визуализация 2: Boxplot для выявления выбросов
if len(numerical_columns) > 0:
    print("\nВизуализация 2: Boxplot для выявления выбросов")
    plt.figure(figsize=(15, 8))
    for i, col in enumerate(numerical_columns, 1):
        plt.subplot(2, 3, i) if len(numerical_columns) > 3 else plt.subplot(1, len(numerical_columns), i)
        df.boxplot(column=col)
        plt.title(f'Boxplot {col}')
    plt.tight_layout()
    plt.show()

# Визуализация 3: Матрица корреляций (только для числовых данных)
if len(numerical_columns) > 1:
    print("\nВизуализация 3: Матрица корреляций")
    plt.figure(figsize=(10, 8))
    correlation_matrix = df[numerical_columns].corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
    plt.title('Матрица корреляций между числовыми переменными')
    plt.show()

    # Вывод корреляций в числовом виде
    print("\nКорреляции между числовыми переменными:")
    display(correlation_matrix)

# Визуализация 4: Scatter plot для связи между двумя переменными
if len(numerical_columns) >= 2:
    print("\nВизуализация 4: Scatter plot")
    # Берем первые два числовых столбца для примера
    col1, col2 = numerical_columns[0], numerical_columns[1]
    plt.figure(figsize=(10, 6))
    plt.scatter(df[col1], df[col2], alpha=0.6, color='green')
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.title(f'Зависимость {col2} от {col1}')
    plt.show()

# Визуализация 5: Количество наблюдений по категориальным переменным
if len(categorical_columns) > 0:
    print("\nВизуализация 5: Распределение категориальных переменных")
    plt.figure(figsize=(15, 10))
    for i, col in enumerate(categorical_columns, 1):
        plt.subplot(2, 2, i) if len(categorical_columns) > 2 else plt.subplot(1, len(categorical_columns), i)
        df[col].value_counts().plot(kind='bar', color='lightcoral')
        plt.title(f'Распределение {col}')
        plt.xlabel(col)
        plt.ylabel('Количество')
        plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Визуализация 6: Связь категориальных и числовых переменных
if len(categorical_columns) > 0 and len(numerical_columns) > 0:
    print("\nВизуализация 6: Связь категориальных и числовых переменных")
    categorical_col = categorical_columns[0]
    numerical_col = numerical_columns[0]

    plt.figure(figsize=(12, 6))
    sns.boxplot(data=df, x=categorical_col, y=numerical_col)
    plt.title(f'Распределение {numerical_col} по категориям {categorical_col}')
    plt.xticks(rotation=45)
    plt.show()

# =============================================================================
# 6. ДОПОЛНИТЕЛЬНЫЙ АНАЛИЗ
# =============================================================================

print("\n6. ДОПОЛНИТЕЛЬНЫЙ АНАЛИЗ:")

# Анализ дубликатов
duplicates = df.duplicated().sum()
print(f"\nКоличество полных дубликатов строк: {duplicates}")

# Анализ уникальных значений
print("\nКоличество уникальных значений в каждом столбце:")
for col in df.columns:
    print(f"{col}: {df[col].nunique()} уникальных значений")

# =============================================================================
# 7. СОХРАНЕНИЕ РЕЗУЛЬТАТОВ
# =============================================================================

print("\n7. СОХРАНЕНИЕ РЕЗУЛЬТАТОВ...")

# Сохранение обработанных данных (опционально)
df.to_csv('processed_data.csv', index=False)
files.download('processed_data.csv')

# Сохранение отчета о данных
with open('data_analysis_report.txt', 'w') as f:
    f.write("ОТЧЕТ ПО АНАЛИЗУ ДАННЫХ\n")
    f.write("=" * 50 + "\n")
    f.write(f"Размерность данных: {df.shape}\n")
    f.write(f"Пропущенные значения:\n{missing_data}\n")
    f.write(f"Типы данных:\n{df.dtypes}\n")

print("Анализ завершен! Файл 'processed_data.csv' готов к скачиванию.")

print("\n" + "="*60)
print("АНАЛИЗ ДАННЫХ УСПЕШНО ЗАВЕРШЕН!")
print("="*60)
```
